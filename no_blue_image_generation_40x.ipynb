# This code tool is for generating cell images without blue nucleus

```python
from torchvision import utils
import os
import shutil
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import tifffile


def create_noblue(source_path, target_path):
    for filename in os.listdir(source_path):
        if filename.lower().endswith((".tif", ".tiff")):
            in_path = os.path.join(source_path, filename)
            out_path = os.path.join(target_path, filename)

            img = tifffile.imread(in_path)

            if img.ndim == 3 and img.shape[-1] == 3:  # (H, W, 3)
                img[:, :, 2] = 0
            elif img.ndim == 3 and img.shape[0] == 3:  # (3, H, W)
                img[2, :, :] = 0
            else:
                print("not image file", filename)
                continue

            tifffile.imwrite(out_path, img, photometric="rgb")
            print("save success!", out_path)
```

```python
import os
# user input the source and target address for cancer and normal cells
source_cancer = "D:\\cell_40x\\cancer"
target_cancer = "D:\\cell_40x_noblue\\cancer"
source_normal = "D:\\cell_40x\\normal"
target_normal = "D:\\cell_40x_noblue\\normal"
os.makedirs(target_normal, exist_ok=True)
os.makedirs(target_cancer, exist_ok=True)
create_noblue(source_cancer, target_cancer)
create_noblue(source_normal, target_normal)
```
